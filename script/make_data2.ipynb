```python
import os

file_list = os.listdir(os.path.join('dataset', 'raw_data', 'more'))
print(file_list)
```

```python
label_data = dict()
labels = [x[:2] for x in file_list]
print(labels)

for l in labels:
    label_data[l] = []

print(label_data)
```

```python
for l in labels:
    path = os.path.join('dataset', 'raw_data', 'more', l + '.txt')
    with open(path, 'r', encoding='utf-8') as f:
        tmp_data = f.readlines()
    tmp_data = [x.strip() for x in tmp_data if x.strip() != ""]
    for item in tmp_data:
        label = l
        label_data[l].append({
            'sentence': item,
            'label': label
        })
print(label_data)
```

```python
file_content = []

for l in labels:
    file_content += label_data[l]
print(len(file_content))
```

```python
import synonyms

addition = []

for sentence_item in file_content:
    seg_output = synonyms.seg(sentence_item['sentence'])
    print(sentence_item['sentence'])
    for i in range(len(seg_output[0])):
        sim_word_list = synonyms.nearby(seg_output[0][i])
        for word, rate in zip(sim_word_list[0], sim_word_list[1]):
            if rate >= 0.7 and rate < 1.0:
                tmp_sentence = ''
                for j in range(len(seg_output[0])):
                    if j == i:
                        tmp_sentence += word
                    else:
                        tmp_sentence += seg_output[0][j]
                # print(tmp_sentence)
                addition.append({
                    'sentence': tmp_sentence,
                    'label': sentence_item['label']
                })

print(len(addition))
```

```python
print(len(addition))
```

```python
file_content += addition

print(len(file_content))
label_cnt = dict()
for l in labels:
    label_cnt[l] = 0

for item in file_content:
    label_cnt[item['label']] += 1

print(label_cnt)
```
